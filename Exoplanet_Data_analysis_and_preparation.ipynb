```python
# Drop columns with all missing values
df = df.drop(columns=[col for col in ['koi_teq_err1', 'koi_teq_err2'] if col in df.columns])


# Fill remaining missing numeric values with median
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Fill categorical missing with 'Unknown'
categorical_cols = df.select_dtypes(include='object').columns
df[categorical_cols] = df[categorical_cols].fillna('Unknown')
```

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = df.select_dtypes(include='number').columns
cols_per_row = 5
rows = math.ceil(len(num_cols)/cols_per_row)

plt.figure(figsize=(20, rows*3))
for i, col in enumerate(num_cols):
    plt.subplot(rows, cols_per_row, i+1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

plt.figure(figsize=(20,10))
for i, col in enumerate(cat_cols):
    plt.subplot(2, 3, i+1)
    # Use y for the categorical column, assign hue=y (same) and hide legend
    sns.countplot(y=df[col], hue=df[col], dodge=False, palette='Set2', legend=False)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,15))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Feature Correlation')
plt.show()
```
