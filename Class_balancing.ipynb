```python
# ======================================================
# balance_RF.ipynb
# ======================================================
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler

csv_path = "/content/drive/MyDrive/Exoplanet/koi_processed_data.csv"
features_path = "/content/drive/MyDrive/Exoplanet/selected_features.txt"

# Load data
df = pd.read_csv(csv_path)
with open(features_path) as f:
    features = [line.strip() for line in f if line.strip()]

X = df[features]
y = df['target']

X_train, _, y_train, _ = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print("Before:", Counter(y_train))

rus = RandomUnderSampler(sampling_strategy=1.0, random_state=42)
X_res, y_res = rus.fit_resample(X_train, y_train)
print("After:", Counter(y_res))

balanced_df = pd.DataFrame(X_res, columns=features)
balanced_df['target'] = y_res
balanced_df.to_csv("/content/drive/MyDrive/Exoplanet/balanced_RF.csv", index=False)
print("✅ balanced_RF.csv saved successfully!")
```

```python
# ======================================================
# balance_XGB.ipynb
# ======================================================
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler

csv_path = "/content/drive/MyDrive/Exoplanet/koi_processed_data.csv"
features_path = "/content/drive/MyDrive/Exoplanet/selected_features.txt"

df = pd.read_csv(csv_path)
with open(features_path) as f:
    features = [line.strip() for line in f if line.strip()]

X = df[features]
y = df['target']

X_train, _, y_train, _ = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print("Before:", Counter(y_train))

ros = RandomOverSampler(sampling_strategy=1.0, random_state=42)
X_res, y_res = ros.fit_resample(X_train, y_train)
print("After:", Counter(y_res))

balanced_df = pd.DataFrame(X_res, columns=features)
balanced_df['target'] = y_res
balanced_df.to_csv("/content/drive/MyDrive/Exoplanet/balanced_XGB.csv", index=False)
print("✅ balanced_XGB.csv saved successfully!")
```

```python
# ======================================================
# balance_NN.ipynb
# ======================================================
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

csv_path = "/content/drive/MyDrive/Exoplanet/koi_processed_data.csv"
features_path = "/content/drive/MyDrive/Exoplanet/selected_features.txt"

df = pd.read_csv(csv_path)
with open(features_path) as f:
    features = [line.strip() for line in f if line.strip()]

X = df[features]
y = df['target']

X_train, _, y_train, _ = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print("Before:", Counter(y_train))

sm = SMOTE(sampling_strategy=1.0, random_state=42, k_neighbors=5)
X_res, y_res = sm.fit_resample(X_train, y_train)
print("After:", Counter(y_res))

balanced_df = pd.DataFrame(X_res, columns=features)
balanced_df['target'] = y_res
balanced_df.to_csv("/content/drive/MyDrive/Exoplanet/balanced_NN.csv", index=False)
print("✅ balanced_NN.csv saved successfully!")
```

```python
# ======================================================
# balance_SVM.ipynb
# ======================================================
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTETomek

csv_path = "/content/drive/MyDrive/Exoplanet/koi_processed_data.csv"
features_path = "/content/drive/MyDrive/Exoplanet/selected_features.txt"

df = pd.read_csv(csv_path)
with open(features_path) as f:
    features = [line.strip() for line in f if line.strip()]

X = df[features]
y = df['target']

X_train, _, y_train, _ = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print("Before:", Counter(y_train))

smt = SMOTETomek(random_state=42)
X_res, y_res = smt.fit_resample(X_train, y_train)
print("After:", Counter(y_res))

balanced_df = pd.DataFrame(X_res, columns=features)
balanced_df['target'] = y_res
balanced_df.to_csv("/content/drive/MyDrive/Exoplanet/balanced_SVM.csv", index=False)
print("✅ balanced_SVM.csv saved successfully!")
```
